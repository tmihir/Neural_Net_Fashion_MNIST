```python
test_data = pd.read_csv("fashion-mnist_test.csv")
test_data.head()
```

```python
test_data.shape
```

```python
y_test = test_data.iloc[:,0]
y_test.head()
```

```python
def BaggingAlgorithm(trainD, testD, number_of_tree):
    
    var = 'Tree'
    model = DecisionTreeClassifier()
    treeDF = pd.DataFrame()
    
    for i in range(number_of_tree):
        
        title = var + " " + str(number_of_tree)
        random_number = random.randint(393,785)
        random_data = trainD.sample(frac = 1.0, replace = True)
        X_train = random_data.iloc[:,1:random_number]
        y_train = random_data.iloc[:,0]
        
        X_test = testD.iloc[:,1:random_number]
        
        model.fit(X_train,y_train)
        y_pred = model.predict(X_test)
        
        treeDF[title] = y_pred
        
    return treeDF
```

```python
tree1 = BaggingAlgorithm(train_data,test_data,1)
cm(y_pred=tree1,y_true=y_test)
```

```python
tree10 = BaggingAlgorithm(train_data,test_data,10)
final_value = tree10.mode(axis=1).iloc[:,0]
cm(final_value,y_test)
```

```python
tree100 = BaggingAlgorithm(train_data,test_data,100)
final_value = tree10.mode(axis=1).iloc[:,0]
cm(final_value,y_test
```

```python
#Neural Network
X_train = train_data.iloc[:,1:].values
X_test = test_data.iloc[:,1:].values

y_train = train_data.iloc[:,0].values
y_train = np_utils.to_categorical(y_train, num_classes=10)
```

```python
model = Sequential()

model.add(Dense(output_dim = 64, init = 'uniform', activation = 'relu', input_dim = 784))

model.add(Dense(output_dim = 32, init = 'uniform', activation = 'relu'))

model.add(Dense(output_dim = 10, init = 'uniform', activation = 'softmax'))

model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics= ['accuracy'])
```

```python
model.fit(X_train,y_train, epochs= 40, batch_size= 128)

y_pred = model.predict(X_test)
```

```python
newDataframe = pd.DataFrame()
for i in range(0,10):
    newDataframe[str(i)] = y_pred[:,i]
    
final_answer = newDataframe.idxmax(axis = 1).astype(int)
```

```python
cm(final_answer, y_test)
```
